```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

# =========================================================
# DATABASE SILVER
# =========================================================

spark.sql("CREATE DATABASE IF NOT EXISTS silver")

# =========================================================
# CLIENTES
# =========================================================

df_clientes = spark.table("bronze.clientes")

df_clientes_silver = (
    df_clientes
    .dropDuplicates(["id_cliente"])
    .withColumn("nome_cliente", initcap(trim(col("nome_cliente"))))
    .withColumn("email", lower(trim(col("email"))))
    .withColumn("telefone", regexp_replace(col("telefone"), "[^0-9]", ""))
    .withColumn("sexo", upper(trim(col("sexo"))))
    .withColumn("renda_mensal", col("renda_mensal").cast("decimal(10,2)"))
)

(
    df_clientes_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.clientes")
)

# =========================================================
# ENDERECOS CLIENTES
# =========================================================

df_enderecos = spark.table("bronze.enderecos_clientes")

df_enderecos_silver = (
    df_enderecos
    .dropDuplicates(["id_endereco"])
    .withColumn("cidade", upper(trim(col("cidade"))))
    .withColumn("estado", upper(trim(col("estado"))))
    .withColumn("bairro", initcap(trim(col("bairro"))))
    .withColumn("cep", regexp_replace(col("cep"), "[^0-9]", ""))
)

(
    df_enderecos_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.enderecos_clientes")
)

# =========================================================
# VENDEDORES
# =========================================================

df_vendedores = spark.table("bronze.vendedores")

df_vendedores_silver = (
    df_vendedores
    .dropDuplicates(["id_vendedor"])
    .withColumn("nome_vendedor", initcap(trim(col("nome_vendedor"))))
    .withColumn("email", lower(trim(col("email"))))
    .withColumn("cidade", upper(trim(col("cidade"))))
    .withColumn("estado", upper(trim(col("estado"))))
)

(
    df_vendedores_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.vendedores")
)

# =========================================================
# CATEGORIAS
# =========================================================

df_categorias = spark.table("bronze.categorias")

df_categorias_silver = (
    df_categorias
    .dropDuplicates(["id_categoria"])
    .withColumn("nome_categoria", initcap(trim(col("nome_categoria"))))
)

(
    df_categorias_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.categorias")
)

# =========================================================
# SUBCATEGORIAS
# =========================================================

df_subcategorias = spark.table("bronze.subcategorias")

df_subcategorias_silver = (
    df_subcategorias
    .dropDuplicates(["id_subcategoria"])
    .withColumn("nome_subcategoria", initcap(trim(col("nome_subcategoria"))))
)

(
    df_subcategorias_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.subcategorias")
)

# =========================================================
# PRODUTOS
# =========================================================

df_produtos = spark.table("bronze.produtos")

df_produtos_silver = (
    df_produtos
    .dropDuplicates(["id_produto"])
    .withColumn("nome_produto", initcap(trim(col("nome_produto"))))
    .withColumn("marca", upper(trim(col("marca"))))
    .withColumn("preco", col("preco").cast("decimal(10,2)"))
    .withColumn("custo", col("custo").cast("decimal(10,2)"))
)

(
    df_produtos_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.produtos")
)

# =========================================================
# PROMOCOES
# =========================================================

df_promocoes = spark.table("bronze.promocoes")

df_promocoes_silver = (
    df_promocoes
    .dropDuplicates(["id_promocao"])
    .withColumn("nome_promocao", initcap(trim(col("nome_promocao"))))
    .withColumn("percentual_desconto", col("percentual_desconto").cast("decimal(5,2)"))
)

(
    df_promocoes_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.promocoes")
)

# =========================================================
# PRODUTOS PROMOCOES
# =========================================================

df_produtos_promocoes = spark.table("bronze.produtos_promocoes")

df_produtos_promocoes_silver = (
    df_produtos_promocoes
    .dropDuplicates(["id_produto_promocao"])
)

(
    df_produtos_promocoes_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.produtos_promocoes")
)

# =========================================================
# PEDIDOS
# =========================================================

df_pedidos = spark.table("bronze.pedidos")

df_pedidos_silver = (
    df_pedidos
    .dropDuplicates(["id_pedido"])
    .withColumn("status_pedido", upper(trim(col("status_pedido"))))
)

(
    df_pedidos_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.pedidos")
)

# =========================================================
# ITENS PEDIDO
# =========================================================

df_itens = spark.table("bronze.itens_pedido")

df_itens_silver = (
    df_itens
    .dropDuplicates(["id_item_pedido"])
    .withColumn("preco_unitario", col("preco_unitario").cast("decimal(10,2)"))
    .withColumn("desconto", col("desconto").cast("decimal(10,2)"))
)

(
    df_itens_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.itens_pedido")
)

# =========================================================
# PAGAMENTOS
# =========================================================

df_pagamentos = spark.table("bronze.pagamentos")

df_pagamentos_silver = (
    df_pagamentos
    .dropDuplicates(["id_pagamento"])
    .withColumn("tipo_pagamento", upper(trim(col("tipo_pagamento"))))
    .withColumn("valor_pagamento", col("valor_pagamento").cast("decimal(10,2)"))
)

(
    df_pagamentos_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.pagamentos")
)

# =========================================================
# ENTREGAS
# =========================================================

df_entregas = spark.table("bronze.entregas")

df_entregas_silver = (
    df_entregas
    .dropDuplicates(["id_entrega"])
    .withColumn("tipo_entrega", upper(trim(col("tipo_entrega"))))
    .withColumn("valor_frete", col("valor_frete").cast("decimal(10,2)"))
)

(
    df_entregas_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.entregas")
)

# =========================================================
# AVALIACOES
# =========================================================

df_avaliacoes = spark.table("bronze.avaliacoes")

df_avaliacoes_silver = (
    df_avaliacoes
    .dropDuplicates(["id_avaliacao"])
    .withColumn("comentario", trim(col("comentario")))
)

(
    df_avaliacoes_silver.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("silver.avaliacoes")
)

print("Camada Silver criada com sucesso.")
```
